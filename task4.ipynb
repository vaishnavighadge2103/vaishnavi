**Arrays in NumPy:-**

ndarray (N-dimensional array) is the main object in NumPy.

```python
import numpy as np

# 1. Create array using np.array()
arr1 = np.array([10, 20, 30, 40, 50])
print("np.array():", arr1)

# 2. Create array using np.zeros()
arr2 = np.zeros(5)
print("np.zeros():", arr2)

# 3. Create array using np.ones()
arr3 = np.ones(5)
print("np.ones():", arr3)

# 4. Create array using np.arange()
arr4 = np.arange(1, 11, 2)
print("np.arange():", arr4)

# 5. Create array using np.linspace()
arr5 = np.linspace(1, 10, 5)
print("np.linspace():", arr5)

# Array Indexing
print("\nFirst Element:", arr1[0])
print("Last Element:", arr1[-1])

# Array Slicing
print("Elements from index 1 to 3:", arr1[1:4])
```

```python
#Question 3
import numpy as np

# Create Arrays
arr1 = np.array([10, 20, 30, 40])
arr2 = np.array([1, 2, 3, 4])

print("Array 1:", arr1)
print("Array 2:", arr2)

# 1. Arithmetic Operations
print("\nAddition:", arr1 + arr2)
print("Subtraction:", arr1 - arr2)
print("Multiplication:", arr1 * arr2)
print("Division:", arr1 / arr2)

# 2. Broadcasting
print("\nBroadcasting (arr1 + 5):", arr1 + 5)
print("Broadcasting (arr1 * 2):", arr1 * 2)

# 3. Element-wise Functions
print("\nSquare Root:", np.sqrt(arr1))
print("Sine:", np.sin(arr2))
print("Exponential:", np.exp(arr2))
```

```python
#Question 4
import numpy as np

# Create a 2D Array
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print("Original Array:")
print(arr)

# 1. Shape and Dimensions
print("\nShape:", arr.shape)
print("Dimensions:", arr.ndim)

# 2. Reshape
reshaped = arr.reshape(3, 2)
print("\nReshaped Array (3x2):")
print(reshaped)

# 3. Flatten
flat = arr.flatten()
print("\nFlatten Array:")
print(flat)

# 4. Transpose
trans = arr.transpose()
print("\nTranspose:")
print(trans)

# 5. Concatenate
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("\nConcatenate:")
print(np.concatenate((a, b)))

# 6. Vertical Stack
print("\nVertical Stack:")
print(np.vstack((a, b)))

# 7. Horizontal Stack
print("\nHorizontal Stack:")
print(np.hstack((a, b)))

# 8. Split Array
c = np.array([10, 20, 30, 40, 50, 60])

print("\nSplit Array:")
print(np.split(c, 3))
```

```python
#Question 5
import numpy as np

# Create a 2D Array
arr = np.array([[10, 20, 30],
                [40, 50, 60],
                [70, 80, 90]])

print("Original Array:")
print(arr)

# 1. Indexing
print("\nFirst Element:", arr[0, 0])
print("Last Element:", arr[2, 2])

# 2. Slicing
print("\nFirst Row:", arr[0])
print("Second Column:", arr[:, 1])
print("Sub Array:")
print(arr[0:2, 1:3])

# 3. Masking (Logical Indexing)
print("\nElements Greater Than 40:")
print(arr[arr > 40])

# 4. Fancy Indexing
a = np.array([10, 20, 30, 40, 50])

print("\nOriginal 1D Array:", a)
print("Fancy Indexing:", a[[0, 2, 4]])
```

```python
#Question 6

import numpy as np

# Create Arrays
a = np.array([10, 20, 30])
b = np.array([1, 2, 3])

print("Array A:", a)
print("Array B:", b)

# 1. Universal Functions (ufuncs)
print("\nAddition:", np.add(a, b))
print("Subtraction:", np.subtract(a, b))
print("Multiplication:", np.multiply(a, b))

# 2. Statistical Functions
arr = np.array([10, 20, 30, 40, 50])

print("\nMean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
print("Variance:", np.var(arr))
print("Minimum:", np.min(arr))
print("Maximum:", np.max(arr))

# 3. Linear Algebra Functions
m1 = np.array([[1, 2],
               [3, 4]])

m2 = np.array([[5, 6],
               [7, 8]])

print("\nDot Product:")
print(np.dot(m1, m2))

print("\nMatrix Multiplication:")
print(np.matmul(m1, m2))

print("\nInverse Matrix:")
print(np.linalg.inv(m1))

print("\nEigen Values and Eigen Vectors:")
values, vectors = np.linalg.eig(m1)
print("Eigen Values:")
print(values)
print("Eigen Vectors:")
print(vectors)
```

**What is Broadcasting:-**

Broadcasting is a feature in NumPy that allows arithmetic operations between arrays of different shapes.
NumPy automatically expands the smaller array to match the shape of the larger array without copying data.

**Rules of Broadcasting:-**

Compare array shapes from right to left.
Two dimensions are compatible if:
      They are equal,
           or
      One of them is 1.
If the dimensions are not compatible, NumPy raises a ValueError.

```python
import numpy as np

# Example 1: Broadcasting with Scalar
arr = np.array([10, 20, 30, 40])

print("Original Array:", arr)
print("Array + 5:", arr + 5)

# Example 2: Broadcasting with 1D Arrays
a = np.array([1, 2, 3])
b = np.array([10, 20, 30])

print("\nArray A:", a)
print("Array B:", b)
print("A + B:", a + b)

# Example 3: Broadcasting with 2D Array
c = np.array([[1], [2], [3]])
d = np.array([10, 20, 30])

print("\n2D Array:")
print(c)

print("\n1D Array:")
print(d)

print("\nBroadcasting Result:")
print(c + d)
```
