```python
#question 1
def maximum(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

print("Maximum number is:", maximum(a, b, c))
```

```python
#question 2
def sum_list(numbers):
    total = 0
    for i in numbers:
        total += i
    return total

numbers = [8, 2, 3, 0, 7]

print("List:", numbers)
print("Sum:", sum_list(numbers))
```

```python
#question 3
def unique_list(numbers):
    unique = []
    for i in numbers:
        if i not in unique:
            unique.append(i)
    return unique

numbers = [1, 2, 3, 3, 3, 3, 4, 5]

print("Original List:", numbers)
print("Unique List:", unique_list(numbers))
```

```python
#question 4
def combinations(items):
    count = 0

    for i in range(len(items)):
        for j in range(i + 1, len(items)):
            print(items[i] + items[j])
            count += 1

    print("Total Combinations:", count)

letters = ["a", "b", "c"]

combinations(letters)
```

```python
#question 5
def permutations(items):
    count = 0

    for i in range(len(items)):
        for j in range(len(items)):
            if i != j:
                print(items[i] + items[j])
                count += 1

    print("Total Permutations:", count)

letters = ["a", "b", "c"]

permutations(letters)
```

```python
#question 6
def count_vowel_consonant(word):
    vowels = "aeiouAEIOU"
    v = 0
    c = 0

    for ch in word:
        if ch.isalpha():
            if ch in vowels:
                v += 1
            else:
                c += 1

    print("Count of vowel is =", v)
    print("Count of consonant is =", c)

word = input("Enter a word: ")
count_vowel_consonant(word)
```

```python
#question 8
def count_case(text):
    upper = 0
    lower = 0

    for ch in text:
        if ch.isupper():
            upper += 1
        elif ch.islower():
            lower += 1

    print("Upper case:", upper)
    print("Lower case:", lower)

text = input("Enter a string: ")
count_case(text)
```

```python
# -------------------- Example 1 --------------------
# Count Uppercase and Lowercase Letters

def count_case(text):
    upper = 0
    lower = 0

    for ch in text:
        if ch.isupper():
            upper += 1
        elif ch.islower():
            lower += 1

    print("Upper case:", upper)
    print("Lower case:", lower)

text = input("Enter a string: ")
count_case(text)


# -------------------- Example 2 --------------------
# Count Vowels and Consonants

def count_vowel_consonant(word):
    vowels = "aeiouAEIOU"
    v = 0
    c = 0

    for ch in word:
        if ch.isalpha():
            if ch in vowels:
                v += 1
            else:
                c += 1

    print("Vowels:", v)
    print("Consonants:", c)

word = input("\nEnter a word: ")
count_vowel_consonant(word)


# -------------------- Example 3 --------------------
# Convert Lowercase to Uppercase

def to_upper(word):
    print("Uppercase:", word.upper())

word = input("\nEnter a lowercase word: ")
to_upper(word)


# -------------------- Example 4 --------------------
# Find Maximum of Three Numbers

def maximum(a, b, c):
    if a >= b and a >= c:
        print("Maximum =", a)
    elif b >= a and b >= c:
        print("Maximum =", b)
    else:
        print("Maximum =", c)

a = int(input("\nEnter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))
maximum(a, b, c)


# -------------------- Example 5 --------------------
# Sum of Numbers in a List

def sum_list(numbers):
    total = 0
    for i in numbers:
        total += i
    print("Sum =", total)

numbers = [8, 2, 3, 0, 7]
sum_list(numbers)
```
