```python
import pandas as pd
import matplotlib.pyplot as plt

s = pd.Series([10, 20, 30, 40], index=['A', 'B', 'C', 'D'])
print(s)

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 22, 35],
    'Marks': [85, 92, None, 95],
    'Department': ['IT', 'HR', 'IT', 'HR']
})

print(df)

print(df['Name'])
print(df.iloc[0:2])
print(df.loc[0:2, ['Name', 'Marks']])

print(df.isnull())

df['Marks'] = df['Marks'].fillna(df['Marks'].mean())
print(df)

print(df.dropna())

df.rename(columns={'Marks': 'Score'}, inplace=True)
print(df)

print(df[df['Age'] > 25])

print(df[(df['Age'] > 25) & (df['Score'] > 90)])

print(df[(df['Age'] > 25) | (df['Score'] > 90)])

print(df.groupby('Department')['Score'].sum())

print(df['Score'].sum())
print(df['Score'].mean())
print(df['Score'].count())

df2 = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'City': ['Delhi', 'Mumbai', 'Pune', 'Chennai']
})

merged = pd.merge(df, df2, on='Name')
print(merged)

concat_df = pd.concat([df, df2], axis=1)
print(concat_df)

df.to_csv("output.csv", index=False)
csv_df = pd.read_csv("output.csv")
print(csv_df)

df.to_excel("output.xlsx", index=False)
excel_df = pd.read_excel("output.xlsx")
print(excel_df)

time_df = pd.DataFrame({
    'Date': ['2024-01-01', '2024-01-15', '2024-02-10'],
    'Sales': [100, 200, 300]
})

time_df['Date'] = pd.to_datetime(time_df['Date'])
time_df = time_df.set_index('Date')
print(time_df.resample('M').sum())

time_df.plot()
plt.show()
```
