```python
# -------------------- Question 1 --------------------
# Swap first and last element of a list

numbers = [10, 20, 30, 40, 50]

print("Original List:", numbers)

temp = numbers[0]
numbers[0] = numbers[-1]
numbers[-1] = temp

print("After Swapping:", numbers)


# -------------------- Question 2 --------------------
# Count length of a string using for loop

text = input("\nEnter a string: ")

count = 0
for ch in text:
    count += 1

print("Length of String:", count)


# -------------------- Question 3 --------------------
# Sum of only non-negative integers

numbers = [1, 4, -5, -20, 10]

sum = 0
for i in numbers:
    if i >= 0:
        sum += i

print("\nList:", numbers)
print("Sum of Non-Negative Numbers:", sum)


# -------------------- Question 4 --------------------
# Factorial of a number

n = int(input("\nEnter a number: "))

fact = 1
for i in range(1, n + 1):
    fact *= i

print("Factorial =", fact)
```

```python
# -------------------- Question 1 --------------------
# Odd and Even using while loop

n = 1

print("Even Numbers:")
while n <= 10:
    if n % 2 == 0:
        print(n, end=" ")
    n += 1

print("\n")

n = 1
print("Odd Numbers:")
while n <= 10:
    if n % 2 != 0:
        print(n, end=" ")
    n += 1


# -------------------- Question 2 --------------------
# Pattern using while loop

print("\n\nStar Pattern:")
i = 1
while i <= 5:
    j = 1
    while j <= i:
        print("*", end=" ")
        j += 1
    print()
    i += 1


# -------------------- Question 3 --------------------
# Create list from 1 to 20 using while loop

list1 = []
i = 1

while i <= 20:
    list1.append(i)
    i += 1

print("\nList 1 to 20:")
print(list1)


# -------------------- Question 4 --------------------
# Create list from 20 to 1 using while loop

list2 = []
i = 20

while i >= 1:
    list2.append(i)
    i -= 1

print("\nList 20 to 1:")
print(list2)


# -------------------- Question 5 --------------------
# break, continue, pass examples

print("\nBreak Example:")
i = 1
while i <= 10:
    if i == 6:
        break
    print(i, end=" ")
    i += 1

print("\n\nContinue Example:")
i = 0
while i < 10:
    i += 1
    if i == 5:
        continue
    print(i, end=" ")

print("\n\nPass Example:")
i = 1
while i <= 5:
    if i == 3:
        pass
    print(i, end=" ")
    i += 1

# -------------------- Example 1 --------------------
# Print numbers from 1 to 10 using while loop

i = 1
while i <= 10:
    print(i)
    i += 1


# -------------------- Example 2 --------------------
# Print even numbers from 1 to 20 using while loop

i = 2
while i <= 20:
    print(i)
    i += 2


# -------------------- Example 3 --------------------
# Find sum of first 10 numbers

i = 1
sum = 0

while i <= 10:
    sum += i
    i += 1

print("Sum =", sum)


# -------------------- Example 4 --------------------
# Factorial using while loop

n = int(input("Enter a number: "))
fact = 1

while n > 0:
    fact *= n
    n -= 1

print("Factorial =", fact)


# -------------------- Example 5 --------------------
# Reverse a number using while loop

num = int(input("Enter a number: "))
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

print("Reversed Number =", rev)
```
