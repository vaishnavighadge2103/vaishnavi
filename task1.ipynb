```python
# question 1

# Integer
a = 20

# Float
b = 5.5

# Complex Number
c = 3 + 4j

print("Integer:", a)
print("Float:", b)
print("Complex Number:", c)

# Arithmetic Operations
print("\nArithmetic Operations:")
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
```

```python
# question 2

# Variables
a = 10
b = 5

# Arithmetic Operators
print("Arithmetic Operators:")
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)

# Comparison Operators
print("\nComparison Operators:")
print("a == b :", a == b)
print("a != b :", a != b)
print("a > b  :", a > b)
print("a < b  :", a < b)

# Logical Operators
print("\nLogical Operators:")
print("(a > 5 and b < 10):", a > 5 and b < 10)
print("(a < 5 or b < 10):", a < 5 or b < 10)

# Assignment Operators
print("\nAssignment Operators:")
x = 10
print("Initial x =", x)

x += 5
print("x += 5  ->", x)

x -= 3
print("x -= 3  ->", x)

x *= 2
print("x *= 2  ->", x)

x /= 4
print("x /= 4  ->", x)
```

```python
# question 3

# Create a list
numbers = [10, 20, 30, 40, 50]
print("Original List:", numbers)

# 1. Accessing Elements
print("\nAccessing Elements:")
print("First Element:", numbers[0])
print("Last Element:", numbers[-1])
print("Slicing (1:4):", numbers[1:4])

# 2. Modifying Elements
numbers[2] = 35
print("\nAfter Modifying:", numbers)

# 3. Adding Elements
numbers.append(60)
print("After append(60):", numbers)

numbers.insert(2, 25)
print("After insert(2, 25):", numbers)

# 4. Removing Elements
numbers.remove(40)
print("After remove(40):", numbers)

numbers.pop()
print("After pop():", numbers)

# 5. Other List Methods
numbers.extend([70, 80])
print("After extend([70, 80]):", numbers)

numbers.sort()
print("After sort():", numbers)

numbers.reverse()
print("After reverse():", numbers)

print("Count of 20:", numbers.count(20))
print("Index of 35:", numbers.index(35))

copy_list = numbers.copy()
print("Copied List:", copy_list)

numbers.clear()
print("After clear():", numbers)
```

```python
# question 4
# Create a tuple
t1 = (10, 20, 30, 40, 20)
print("Original Tuple:", t1)

# 1. Accessing Elements
print("\nAccessing Elements:")
print("First Element:", t1[0])
print("Last Element:", t1[-1])

# 2. Slicing
print("\nSlicing:")
print("Tuple[1:4]:", t1[1:4])

# 3. Concatenation
t2 = (50, 60)
t3 = t1 + t2
print("\nAfter Concatenation:", t3)

# 4. Repetition
print("\nRepetition:")
print(t2 * 3)

# 5. Count Method
print("\nCount:")
print("Count of 20:", t1.count(20))

# 6. Index Method
print("\nIndex:")
print("Index of 30:", t1.index(30))
```

```python
# Question 5

# Create two sets
set1 = {10, 20, 30, 40}
set2 = {30, 40, 50, 60}

print("Set 1:", set1)
print("Set 2:", set2)

# 1. Add an element
set1.add(50)
print("\nAfter add(50):", set1)

# 2. Remove an element
set1.remove(20)
print("After remove(20):", set1)

# 3. Union
print("\nUnion:", set1.union(set2))

# 4. Intersection
print("Intersection:", set1.intersection(set2))

# 5. Difference
print("Difference (set1 - set2):", set1.difference(set2))
print("Difference (set2 - set1):", set2.difference(set1))
```

```python
# Question 6

# Create a dictionary
student = {
    "Name": "Vedant",
    "Age": 20,
    "Course": "Computer Engineering"
}

print("Original Dictionary:", student)

# 1. Accessing Key-Value Pairs
print("\nAccessing Values:")
print("Name:", student["Name"])
print("Age (using get):", student.get("Age"))

# 2. Adding a New Item
student["City"] = "Pune"
print("\nAfter Adding City:", student)

# 3. Updating an Item
student.update({"Age": 21})
print("After Updating Age:", student)

# 4. Keys
print("\nKeys:", student.keys())

# 5. Values
print("Values:", student.values())

# 6. Items
print("Items:", student.items())

# 7. Removing an Item
student.pop("Course")
print("\nAfter pop('Course'):", student)
```

```python
#Question 7
# Cause: Incorrect indentation after if statement.
# Wrong:
# if True:
# print("Hello")

# Correct:
if True:
    print("IndentationError Fixed")


# 2. NameError
# Cause: Using a variable before defining it.
# Wrong:
# print(x)

# Correct:
x = 10
print("NameError Fixed:", x)


# 3. ValueError
# Cause: Invalid value for a function.
# Wrong:
# num = int("abc")

# Correct:
try:
    num = int("123")
    print("ValueError Fixed:", num)
except ValueError:
    print("Invalid Value")


# 4. TypeError
# Cause: Unsupported operation between different data types.
# Wrong:
# result = "10" + 5

# Correct:
result = int("10") + 5
print("TypeError Fixed:", result)


# 5. IndexError
# Cause: Accessing an index that does not exist.
numbers = [10, 20, 30]

try:
    print(numbers[2])      # Valid index
    print(numbers[5])      # Invalid index
except IndexError:
    print("IndexError: List index out of range")


# 6. KeyError
# Cause: Accessing a key that does not exist in a dictionary.
student = {"name": "Vedant", "age": 20}

try:
    print(student["name"])     # Existing key
    print(student["city"])     # Non-existing key
except KeyError:
    print("KeyError: Key not found")
```
